```python
from bs4 import BeautifulSoup
import requests
import time
import numpy as np
import pandas as pd
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/102.0.0.0 Safari/537.36'}
url_token = 'https://rent.591.com.tw/?region=1&order=posttime&orderType=desc'
res = []

with requests.session() as sess:
    #get token

    html = sess.get(url_token, headers=headers).text
    soup = BeautifulSoup(html, 'html.parser')
    x_csrf_token = soup.select_one('meta[name="csrf-token"]')['content']
    
    time.sleep(np.random.random()*3)
    print(x_csrf_token)
    
    new_headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/102.0.0.0 Safari/537.36', 
    'X-CSRF-TOKEN': x_csrf_token,
    'X-Requested-With': 'XMLHttpRequest'
    }
    
    page_num = 0
    while True:
        url = f'https://rent.591.com.tw/home/search/rsList?is_format_data=1&is_new_list=1&type=1&region=1&order=posttime&orderType=desc&firstRow={page_num}'
        data = sess.get(url, headers=new_headers)
        posts = data.json()['data']['data']
        
        if not posts:
            break
        for post in posts:
            df = pd.json_normalize(post, sep='_')
            res.append(df)
        break
        page_num += 30
        
        time.sleep(np.random.random())
        
```

```python
res = pd.concat(res)
res
```

```python
res.to_csv('591_TaiChung.csv', index=False, encoding='utf_8_sig')
```
